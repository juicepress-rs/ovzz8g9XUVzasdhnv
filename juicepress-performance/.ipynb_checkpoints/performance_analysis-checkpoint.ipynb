```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
from pathlib import Path
from matplotlib import rc


```

### NOTE: UNPACK THE RESULTS FOLDER BEFORE RUNNING THIS NOTEBOOK (`7z x reports.7z`)

```python
rc("font", **{"family": "serif", "serif": ["Times"], "size": 15})
rc("text", usetex=False)
```

```python
BASE = Path("juicepress_results_corpus_j/reports")
```

```python
raw = {
    "tree_id": [],
    "total_sec": [],
    "analysis_sec": [],
    "collection_sec": [],
    "file_count": [],
    "elf_cum_size": [],
    "elf_count": [],
    "tokio_sec": [],
    "rayon_sec": [],
    "total_sec_sequential": [],
}

rayon_keys = set(["keyword_in_binary", "known_taint_symbol", "keyword_in_filename"])

for i, p_report in enumerate(BASE.glob("*juicepress-masked.json")):
    print(i)
    raw["tree_id"] += [p_report.stem.split("___")[2]]

    data = json.loads(p_report.read_text())
    
    for k in raw.keys() - {"tree_id", "tokio_sec", "rayon_sec", "total_sec_sequential", "elf_cum_size"}:
        raw[k] += [data[k]]
    rayon_sec = 0.0
    tokio_sec = 0.0
    elf_cum_size = 0
    for result in data["results"]:
        elf_cum_size += result["size"]
        rbf = result["ranking_by_factor"]
        rayon_sec += sum(rbf[k]["exec_secs"] for k in rayon_keys)
        tokio_sec += sum(rbf[k]["exec_secs"] for k in rbf.keys() - rayon_keys)
    raw["elf_cum_size"] = elf_cum_size
    raw["rayon_sec"] += [rayon_sec]
    raw["tokio_sec"] += [tokio_sec]
    raw["total_sec_sequential"] += [tokio_sec + rayon_sec]
```

```python
df = pd.DataFrame(raw)
df["tokio_share"] = df["tokio_sec"] / df["total_sec_sequential"]
df["rayon_share"] = df["rayon_sec"] / df["total_sec_sequential"]
df["time_per_file"] = df["total_sec"] / df["file_count"]
df["time_per_elf"] = df["total_sec"] / df["elf_count"]
df["size_per_elf"] = df["elf_cum_size"] / df["elf_count"]
df
```

```python
from matplotlib.ticker import ScalarFormatter, FuncFormatter
from matplotlib.colors import LinearSegmentedColormap
plt.figure(figsize=(8, 5.5)) 

palette = sns.color_palette("colorblind")
blue = palette[0]    # blue
orange = palette[1]  # orange
yellow = "#ECE133"
red = palette[3]     # red from colorblind palette

# Create continuous colormap: blue -> orange -> red
cmap = LinearSegmentedColormap.from_list("blue_orange_red_yellow", [blue, red, orange, yellow])

ymax = 100


x_p90 = df["elf_count"].quantile(0.90)
x_p95 = df["elf_count"].quantile(0.95)
x_p99 = df["elf_count"].quantile(0.99)

# P95 and P99 for y-axis (total_sec)
y_p90= df["total_sec"].quantile(0.90)
y_p95 = df["total_sec"].quantile(0.95)
y_p99 = df["total_sec"].quantile(0.99)

ax = sns.histplot(
    data=df,
    x="elf_count",
    y="total_sec",
    bins=10,           # number of bins for both axes
    pmax=0.95,          # optional: cap extreme values for color scaling
    cbar=True,          # show colorbar
    cmap=cmap,               # color map
    log_scale=(True, True),  # log-log scale
    edgecolor="black",  # border color for each patch
    linewidth=0.5,       # thickness of the border
)



for xp, yp, offs, label, style in zip([x_p95, x_p99], [y_p95, y_p99], [(-70.0, -3), (-250.0, -6)], [f"P95", f"P99$\\setminus$P95"], [":", ":"]):
    ax.plot([xp, xp], [0.000001, yp], color="black", linestyle="--", linewidth=2)
    ax.plot([0.000001, xp], [yp, yp], color="black", linestyle="--", linewidth=2)
    ax.plot([xp, xp], [yp, ymax], color="grey", linestyle=":", linewidth=2)
    ax.plot([xp, 4000.00], [yp, yp], color="grey", linestyle=":", linewidth=2)
    ax.text(xp + offs[0], yp + offs[1], label, color="black", fontweight="bold", fontsize=20,
            ha="right", va="top")

ax.text(2500, 52, "$>$P99", color="black", fontweight="heavy", fontsize=20, ha="right", va="top")


ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{x:.0f}"))
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f"{y:.2f}"))
ax.set_ylabel("Exec. Time per File Tree [s, log]")
ax.set_xlabel("ELF Files in File Tree [\\#, log]")

current_ticks = list(ax.get_yticks())
current_ticks.append(y_p99)
ax.set_yticks(current_ticks) 
current_labels = [f"{tick:.2f}" for tick in current_ticks]
ax.set_yticklabels(current_labels)

current_ticks = list(ax.get_xticks())
print(current_ticks)

current_ticks.append(100.)
current_ticks.append(x_p95)
current_ticks.append(x_p99)
ax.set_xticks(current_ticks) 
current_labels = [f"{tick:.0f}" for tick in current_ticks]
ax.set_xticklabels(current_labels)

cbar = ax.collections[0].colorbar.set_label("Processed File Trees in Bin [\\#]")

ax.set_ylim(0.015, 100)
ax.set_xlim(2, 2800)


plt.tight_layout()
plt.savefig("juicepress_perf_eval.pdf", bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(8,5.5))

ax = sns.kdeplot(data=df["rayon_share"], fill=True, alpha=1, color=palette[0], zorder=1, label="CPU-Bound Factors")
sns.kdeplot(df["rayon_share"], color="black", linewidth=2, fill=False, zorder=1, ax=ax, label=None)
sns.kdeplot(data=df["tokio_share"], fill=True, alpha=0.9, color=palette[1], zorder=2, label="IO-Bound Factors", ax=ax)
sns.kdeplot(df["tokio_share"], color="black", linewidth=2, fill=False, ax=ax, zorder=2, label=None)

ax.set_axisbelow(True)
ax.grid(True, zorder=-5)

ax.legend(ncols=2, loc="upper center", bbox_to_anchor=(0.5, 1.115))

#title="Keyword Matchers",
ax.set_xlabel("Share on Cumulative Execution Time per File Tree Scoring")
plt.savefig("execution_share_per_file_tree_scoring.pdf", bbox_inches="tight")
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
p = df[df["total_sec"] < y_p95]["total_sec"]
params = stats.gamma.fit(df["total_sec"])

fig = plt.figure(figsize=(8,5.5))
osm, osr = stats.probplot(df["total_sec"], dist="gamma", sparams=params, plot=plt)

ax = fig.get_axes()[0]

# Change scatter points
ax.get_lines()[0].set_markerfacecolor(blue)
ax.get_lines()[0].set_alpha(0.70)
ax.get_lines()[0].set_markeredgecolor(blue)
ax.get_lines()[0].set_rasterized(True)
ax.get_lines()[0].set_label("Execution Time")


# Change diagonal/reference line
ax.get_lines()[1].set_color(orange)
ax.get_lines()[1].set_zorder(1)
ax.get_lines()[1].set_linewidth(4)

ax.set_title(None)
ax.set_ylabel("Ordered Execution Times [s]")
ax.set_xlabel("Theoretical Quantiles")
ax.get_lines()[1].set_label("Gamma fit ($\\alpha \\approx 0.05883,$ loc $\\approx 0.01490$, $\\theta \\approx 2.08047)$")

print(params)

# P95
plt.axhline(y=9.37, color='black', linestyle='--', linewidth=2)
ax.text(12, 6, "P95=9.37", color="black", fontweight="bold", fontsize=15,
            ha="right", va="top")

# P99
plt.axhline(y=22.67, color='black', linestyle='--', linewidth=2)
ax.text(12.22, 19, "P99=22.67", color="black", fontweight="bold", fontsize=15,
            ha="right", va="top")

ax.set_axisbelow(True)
ax.grid(True, zorder=-5)

ax.legend(ncols=1, loc="upper center", bbox_to_anchor=(0.5, 1.1925))

plt.savefig("exec_times_gamma_fit.pdf", bbox_inches="tight")

plt.show()
```
